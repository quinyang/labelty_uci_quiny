```python
import os

# 1. Define the absolute paths to the directories we created earlier
# Replace 'quinyang' with your actual Ubuntu username if it is different
base_dir = "/home/quinyang/labelty_uci_quiny/nnUNet_data"

# 2. Inject them directly into the environment for this notebook session
os.environ['nnUNet_raw'] = os.path.join(base_dir, "nnUNet_raw")
os.environ['nnUNet_preprocessed'] = os.path.join(base_dir, "nnUNet_preprocessed")
os.environ['nnUNet_results'] = os.path.join(base_dir, "nnUNet_results")

print("Environment variables linked successfully!")
print(f"Raw Data Path: {os.environ['nnUNet_raw']}")
```

```python
import subprocess

# Grab the current environment variables (including the nnUNet paths we set)
current_env = os.environ.copy()

print("Starting Plan and Preprocess. This may take a few minutes...")

try:
    subprocess.run(
        "nnUNetv2_plan_and_preprocess -d 101 --verify_dataset_integrity", 
        shell=True, 
        check=True,
        env=current_env
    )
    print("✅ Preprocessing completed successfully!")
except subprocess.CalledProcessError as e:
    print(f"❌ An error occurred during preprocessing: {e}")
```

```python
print("Starting nnU-Net Training (Fold 0, 2D)...")

try:
    # This will run for a long time! The output will stream directly below the cell.
    subprocess.run(
        "nnUNetv2_train 101 2d 0 -c", 
        shell=True, 
        check=True,
        env=current_env
    )
    print("✅ Training completed successfully!")
except subprocess.CalledProcessError as e:
    print(f"❌ An error occurred during training: {e}")
```
